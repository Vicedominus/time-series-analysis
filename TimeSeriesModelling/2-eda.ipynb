```python
df['2016-09-14'].isnull().sum()
```

```python
df_temp = df['2016-09-14'][['TEMP']]
```

```python
df_temp['TEMP_FF'] = df_temp.fillna(method='ffill')
df_temp
```

```python
df_temp['TEMP_BF'] = df_temp['TEMP'].fillna(method='bfill')
df_temp
```

```python

df_temp['TEMP_R5'] = df_temp['TEMP'].fillna(df['TEMP'].rolling(window=5, min_periods=1).mean())
df_temp
```

```python

df_temp['TEMP_R5V'] = df_temp['TEMP'].fillna((df['TEMP'].rolling(window=3, min_periods=1).mean()+df['TEMP'].shift(-3).rolling(window=3, min_periods=1).mean())/2)
df_temp
```
